````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(cv2.__version__)
````

````python
# Reading an image
# list out ways to inspect an image long term goal 

import cv2

# read image using imread() function
image = cv2.imread('image.jpg')

# extract the height and the width of the image
h, w, r = image.shape[:3] # height, width, and number of channels for the image

# display the height wdith channels
print(f"Height = {h}, Width = {w}, channels = {r}") 
````

````python
Row = 100
Column = 100

(B, G, R) = image[Row, Column] # [] used to acces elements in in numpy array

print(f"Blue = {B}, Green = {G}, Red = {R} at Row: {Row} Column: {Column}")
````

````python
import matplotlib.pyplot as plt

channel0= image[:,:,0].copy()
channel1= image[:,:,1].copy()
channel2= image[:,:,2].copy()

combined = cv2.hconcat([channel0, channel1, channel2])



plt.imshow(combined, cmap="gray")
plt.axis("off")
plt.show()
````

````python
plt.imshow(image)
plt.axis("off")
plt.show()
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

blue = np.zeros_like(image)
green = np.zeros_like(image)
red = np.zeros_like(image)

blue[:, :, 0] = image[:, :, 0]
green[:, :, 1] = image[:, :, 1]
red[:, :, 2] = image[:, :, 2]

combined = cv2.hconcat([blue, green, red])

plt.imshow(cv2.cvtColor(combined, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
````

````python
# matplotlib assume RGB not BGR
# opencv stores data as BGR

plt.imshow(red)
plt.axis("off")
plt.show()
````

````python
mat_red = cv2.cvtColor(red, cv2.COLOR_BGR2RGB)

plt.imshow(mat_red)
plt.axis("off")
plt.show()
````
